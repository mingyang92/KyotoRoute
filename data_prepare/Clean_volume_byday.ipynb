```python
import pandas as pd
import time
```

```python
# read in the data
df_1811 = pd.read_csv('G:/My Drive/Researcher_KyotoU/walkInKyotoGPS_201809-201902/DataOutput/DataOutput/android_1811.csv')
```

```python
# subset unused columns
data = df_1811.drop(columns=['Unnamed: 0','cd4','cd5','cd6','cd7','cd8','cd9','cd10', 'dataVolume', 'IDvolume'])
data.head()
```

```python
data.shape
```

```python
def data_count(df, days=30):
    """df = data of gps points for each month
       days = numbers of days with in that month, default value is 30
       result of the function: create a dataframe, which contains user ID and numbers of data for each month + total data volume.
       Example:
           data_out = data_count(df, 30)
    """
    #create empty dataframe with column names 'uid' and days from 1 - 30
    col_list = [str(x) for x in range(1,days+1)]
    cnames = ['uid'] + col_list + ['total']
    df_out = pd.DataFrame(columns= cnames)

    index = 0
    for uid in df['uid'].unique():
        #create empty list with 30 zeros to store data counts for each day
        count = [0]*(days+1)
        dataID = df[df.uid == uid]
        count[-1] = len(dataID)
        for i in dataID['day']:
            count[i-1] += 1
        temp = [uid] + count
        df_out.loc[index] = temp
        index += 1
        print("still running, index: ", index)
    return df_out
```

```python
start_time = time.time()
out = data_count(data, 30)
print("---Running time: %s seconds ---" % (time.time() - start_time))
```

# Create heatmap for sorted dataframe

```python
# out is the data without sorting, data_des is the data with sorting
data_test = out.iloc[0:,1:31].head(100)
data_test
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(
                    z=data_test,
                    x=[str(x) for x in range(1,31)],
                    showscale=True,
                    #colorscale: RdBu, Viridis, Magma, Cividis,Electric 
                    #colorscale='Electric'
                    colorscale=[[0, "rgb(255,255,255)"],
                                [0.10, "rgb(255,255,255)"],
                                [0.20, "rgb(153,204,255)"],
                                [0.45, "rgb(255,153,51)"],
                                [0.65, "rgb(51,160,44)"],
                                [0.85, "rgb(251,154,153)"],
                                [1, "rgb(227,26,28)"]]
))
fig.update_layout(width=900, height=800,
    title='Number of data points per day',
    xaxis={"title": "Day"}, yaxis={"title": "user"},
    xaxis_nticks=36)

fig.show()
```
