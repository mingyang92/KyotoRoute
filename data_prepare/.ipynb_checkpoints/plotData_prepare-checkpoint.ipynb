# Prepare the data for plotting

```python
import pandas as pd
import time
```

```python
# read in the data
start_time = time.time()
df = pd.read_csv('android_1809.csv')
print("---Running time: %s seconds ---" % (time.time() - start_time))
```

```python
# print the first 5 rows of data
print(df.head())
print(type(df))
```

```python
# subset unused columns
data = df.drop(columns=['Unnamed: 0','cd4','cd5','cd6','cd7','cd8','cd9','cd10', 'dataVolume', 'IDvolume'])
```

```python
data.head()
```

```python
# move certain column to the designated position
def move_column_inplace(df, col, pos):
    col = df.pop(col)
    df.insert(pos, col.name, col)
    print(df.head())
```

```python
# move the 'uid' column to the front
move_column_inplace(data, 'uid', 0)
```

# Convert data to dictionary and sort by uid counts

```python
data.shape
```

```python
start_time = time.time()
# took the first 100 rows of data
td = data
print(td.columns[2:])
# sort the unique uid based on their counts (Higher to lower counts)
td = td.iloc[td.groupby('uid').uid.transform('size').mul(-1).argsort(kind='mergesort')]
# store unique uid names to 'name_df'
name_df = td['uid'].unique()
print(name_df.shape, type(name_df))
print("---Running time: %s seconds ---" % (time.time() - start_time))
```

```python
td.iloc[3]
```

```python
start_time = time.time()
# get length of rows for each unique uid
nameList_uniq = name_df.tolist()
# output uid and length of rows for each uid to dataframe
cols = ['uid', 'length']
lst = []
for a in nameList_uniq:
    lst.append([a,len(td[td['uid']==a])])
df_nameList = pd.DataFrame(lst, columns=cols)
print(df_nameList.head(10))
print("---Running time: %s seconds ---" % (time.time() - start_time))
```

```python
# convert dataframe to dictionary by 'uid'
import collections
start_time = time.time()
nested_dict = lambda: collections.defaultdict(nested_dict)
d = nested_dict()
colList = td.columns[2:]
print(colList)
for i in name_df:
    for ind in td[td['uid']==i].index:
        for col in colList:
            index = td['index'][ind]
            d[i][index][col]=td[col][ind]
print("---Running time: %s seconds ---" % (time.time() - start_time))
```

```python
import pprint
k = 0
for item in d:
    print(item)
    print(len(d[item]))
    pprint.pprint(d[item])
    k = k + 1 
    if(k>1): break
```
