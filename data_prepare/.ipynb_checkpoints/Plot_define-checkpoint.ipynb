```python
import pandas as pd
import time
```

```python
df = pd.read_csv('export_td.csv')
```

```python
df.head()
```

```python
d=[]
for i in range(len(df)):
    temp = df['pos'][i][1:-1].split(',')
    d.append([temp[0],temp[1]])
d_df = pd.DataFrame(d, columns = ['longitude' , 'latitude'])
d_df
```

```python
result = pd.concat([df, d_df], axis=1, sort=False)
result.head()
```

```python
result.to_csv('export_plot.csv', index=False)
```
